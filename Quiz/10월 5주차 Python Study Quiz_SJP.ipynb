### Quiz2. ramdom함수를 이용하여 만든 두개의  표준 정규 분포(-1<range<1) array를 육각형의 분포로 시각화한 것은?

```python
# 1. -1<x,y<1 array를 kde로 시각화
n=10000
df = pd.DataFrame({'x': np.random.randn(n),'y': np.random.randn(n)})
df.plot(kind="kde")
```

```python
#2. 0<x,y<1 array를 hexbin으로 시각화
df = pd.DataFrame({'x': np.random.rand(n),'y': np.random.rand(n)})
df.plot.hexbin(x='x', y='y', gridsize=20)
```

```python
# 3. 정답 -1<x,y<1 array를 hexbin으로 시각화
df = pd.DataFrame({'x': np.random.randn(n),'y': np.random.randn(n)})
df.plot.hexbin(x='x', y='y', gridsize=20)
```

```python
#4. -1<x,y<1 array를 히스토그램으로 시각화
df = pd.DataFrame({'x': np.random.randn(n),'y': np.random.randn(n)})
df.plot.hist()
```
