```python
# 1: map 함수를 이용하여 
df = pd.read_csv('http://bit.ly/ds-korean-idol')
my_map = {'남자':1, '여자':0}
df['성별'] = df['성별'].map(my_map)
df
```

```python
# 3: 성별에 대한 람다 함수를 정의
df = pd.read_csv('http://bit.ly/ds-korean-idol')
f = lambda x:1 if x=='남자' else 0
df['성별']=df['성별'].apply(f)
df
```

```python
# 4: 성별에 대한 함수를 apply
df = pd.read_csv('http://bit.ly/ds-korean-idol')
def male_or_female(x):
    if x == '남자':
        return 1
    elif x == '여자':
        return 0
df = pd.read_csv('http://bit.ly/ds-korean-idol')
df['성별_NEW'] = df['성별'].apply(male_or_female)
df
```
